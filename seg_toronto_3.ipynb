## loading libraries

```python

import requests
from bs4 import BeautifulSoup
import pandas as pd
from IPython.display import display_html
import folium # plotting library
from sklearn.cluster import KMeans
```

## Grap Html table from wikipedia url

```python
ur='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
r = requests.get(ur)
print(r.status_code)
r_text=r.text
```

```python
display_html(r_text,raw=True)
```

## parese html with beautifulsoup

```python
soup=BeautifulSoup(r_text,'html.parser')
tdata=soup.findAll('td')
```

```python
table_contents=[]
table=soup.find('table')
for row in table.findAll('td'):
    cell = {}
    if row.span.text=='Not assigned':
        pass
    else:
        cell['PostalCode'] = row.p.text[:3]
        cell['Borough'] = (row.span.text).split('(')[0]
        cell['Neighborhood'] = (((((row.span.text).split('(')[1]).strip(')')).replace(' /',',')).replace(')',' ')).strip(' ')
        table_contents.append(cell)

# print(table_contents)
df=pd.DataFrame(table_contents)
df['Borough']=df['Borough'].replace({'Downtown TorontoStn A PO Boxes25 The Esplanade':'Downtown Toronto Stn A',
                                             'East TorontoBusiness reply mail Processing Centre969 Eastern':'East Toronto Business',
                                             'EtobicokeNorthwest':'Etobicoke Northwest','East YorkEast Toronto':'East York/East Toronto',
                                             'MississaugaCanada Post Gateway Processing Centre':'Mississauga'})
```

## load Geospatial Coordinates from csv file

```python
g=pd.read_csv('Geospatial_Coordinates.csv')
```

```python
## rename column postal code  to PostalCode to unify both dataset (prepare for merge)
```

```python
g = g.rename(columns={"Postal Code":"PostalCode"})
```

## merge Geospatial Coordinates to our dataset

```python
dfg=pd.merge(df,g,on=['PostalCode'],how='inner')
dfg.head()
```

## New dataset for toronto Boroughs

```python
dftoronto = dfg[dfg['Borough'].str.contains('Toronto',regex=False)]
dftoronto.head()
```

## Getting the center of the map

```python
latMean=dftoronto.Latitude.mean()
lonMean=dftoronto.Longitude.mean()
```

## Drawing the map of toronto with Borough locations

```python
map_toronto = folium.Map(location=[latMean,lonMean],zoom_start=11.5)

for lat,lng,borough,neighbourhood in zip(dftoronto['Latitude'],dftoronto['Longitude'],dftoronto['Borough'],dftoronto['Neighborhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='Blue',
    fill=True,
    fill_color='white',
    fill_opacity=0.5,
    parse_html=False).add_to(map_toronto)
map_toronto
```

## clustering

```python
k=5
kmeans = KMeans(n_clusters = k,random_state=0).fit(dftoronto.loc[:,['Latitude','Longitude']])
kmeans.labels_
dftoronto['lable']=kmeans.labels_
```

## Drawing the map of toronto with Borough locations and colored labels(cluster)

```python
color=['red','blue','green','black','yellow']
map_toronto = folium.Map(location=[latMean,lonMean],zoom_start=11.5)

for lat,lng,borough,neighbourhood,lable in zip(dftoronto['Latitude'],dftoronto['Longitude'],dftoronto['Borough'],dftoronto['Neighborhood'],dftoronto['lable']):
   
    label = '{},\n {},{}'.format(neighbourhood,lat,lng )
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color=color[lable],
    fill=True,
    fill_color=color[lable],
    fill_opacity=0.3,
    parse_html=False).add_to(map_toronto)
map_toronto
```
